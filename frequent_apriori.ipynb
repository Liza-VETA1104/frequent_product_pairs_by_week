```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
#–ü—Ä–∏–º–µ—Ä –¥–∞–Ω–Ω—ã—Ö: —Å–ø–∏—Å–æ–∫ —á–µ–∫–æ–≤ 
transactions = [
    ['–º–æ–ª–æ–∫–æ', '—Ö–ª–µ–±', '–º–∞—Å–ª–æ', '—è–π—Ü–∞', '—á–∞–π'],
    ['–º–æ–ª–æ–∫–æ', '–º–∞—Å–ª–æ', '–∫–æ—Ñ–µ'],
    ['—Ö–ª–µ–±', '–ø–∏–≤–æ', '—á–∏–ø—Å—ã', '–∫–æ–ª–±–∞—Å–∞'],
    ['–º–æ–ª–æ–∫–æ', '—Ö–ª–µ–±', '–º–∞—Å–ª–æ', '–ø–∏–≤–æ', '—á–∏–ø—Å—ã', '—è–π—Ü–∞', '—á–∞–π', '–∫–æ—Ñ–µ'],
    ['–ø–∏–≤–æ', '—á–∏–ø—Å—ã', '–æ—Ä–µ—Ö–∏'],
    ['–º–æ–ª–æ–∫–æ', '—è–π—Ü–∞', '—Å—ã—Ä', '—Ö–ª–µ–±'],
    ['–º–∞—Å–ª–æ', '—Å—ã—Ä', '–∫–æ–ª–±–∞—Å–∞', '—Ö–ª–µ–±'],
    ['–º–æ–ª–æ–∫–æ', '—Ö–ª–µ–±', '–º–∞—Å–ª–æ', '—Å—ã—Ä', '—è–π—Ü–∞', '—á–∞–π', '–æ—Ä–µ—Ö–∏', '–ø–∏–≤–æ', '—á–∏–ø—Å—ã'],
    ['–∫–æ—Ñ–µ', '–º–æ–ª–æ–∫–æ', '—à–æ–∫–æ–ª–∞–¥'],
    ['–ø–∏–≤–æ', '—á–∏–ø—Å—ã', '–º–∞—Å–ª–æ', '—Ö–ª–µ–±']
]

print(f"‚úÖ –ó–∞–≥—Ä—É–∂–µ–Ω–æ {len(transactions)} —á–µ–∫–æ–≤\n")
```

```python
# –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –≤ –±–∏–Ω–∞—Ä–Ω—É—é —Ç–∞–±–ª–∏—Ü—É (—á–µ–∫ * —Ç–æ–≤–∞—Ä)
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

print("üìä –ü—Ä–∏–º–µ—Ä –¥–∞–Ω–Ω—ã—Ö (–ø–µ—Ä–≤—ã–µ 5 —á–µ–∫–æ–≤):")
print(df.head().astype(int))  # –ø–æ–∫–∞–∑—ã–≤–∞–µ–º –∫–∞–∫ 0/1
print("\n" + "-"*50 + "\n")
```

```python
# –ø–æ–∏—Å–∫ —á–∞—Å—Ç—ã—Ö –Ω–∞–±–æ—Ä–æ–≤
min_support = 0.3  # –Ω–∞–±–æ—Ä –¥–æ–ª–∂–µ–Ω –≤—Å—Ç—Ä–µ—á–∞—Ç—å—Å—è –º–∏–Ω–∏–º—É–º –≤ 30% —á–µ–∫–æ–≤

frequent_itemsets = apriori(
    df,
    min_support=min_support,
    use_colnames=True)

frequent_itemsets['size'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))  #—Å–∫–æ–ª—å–∫–æ —Ç–æ–≤–∞—Ä–æ–≤ –≤ –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏
frequent_itemsets = frequent_itemsets.sort_values('support', ascending=False)

print(f"üîç –ù–∞–π–¥–µ–Ω–æ {len(frequent_itemsets)} —á–∞—Å—Ç—ã—Ö –Ω–∞–±–æ—Ä–æ–≤ (min_support = {min_support})\n")
```

```python
# –æ—Å—Ç–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ –ø–∞—Ä—ã —Ç–æ–≤–∞—Ä–æ–≤
pairs = frequent_itemsets[frequent_itemsets['size'] == 2]

if not pairs.empty:
    for i, row in pairs.iterrows():
        items = ', '.join(sorted(row['itemsets']))
        support_percent = row['support'] * 100
        print(f'{items} –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –≤ {support_percent:.1f}% —á–µ–∫–æ–≤')
else:
    print("‚ö†Ô∏è  –ù–µ—Ç –ø–∞—Ä, —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—é—â–∏—Ö min_support.")        
```
